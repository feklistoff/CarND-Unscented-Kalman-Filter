```python
import plotly.offline as py
from plotly.graph_objs import *
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read csv logs
columns = ['NIS_LIDAR', 'NIS_RADAR']
rows = pd.read_csv('./nis.csv', names=columns)
lidar = rows['NIS_LIDAR'].tolist()
radar = rows['NIS_RADAR'].tolist()
```

```python
rows.NIS_LIDAR.describe()
```

```python
rows.NIS_RADAR.describe()
```

```python
plt.figure(figsize=(16, 8))
plt.plot(lidar, color='chocolate')
plt.axhline(y=5.991)
plt.title('NIS LIDAR')
plt.ylabel('NIS')
plt.xlabel('timestep')
plt.show()
```

```python
plt.figure(figsize=(16, 8))
plt.plot(radar[2:], color='chocolate')
plt.axhline(y=7.815)
plt.title('NIS RADAR')
plt.ylabel('NIS')
plt.xlabel('timestep')
plt.show()
```
